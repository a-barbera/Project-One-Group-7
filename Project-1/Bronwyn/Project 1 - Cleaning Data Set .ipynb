```python
# Dependencies and Setup
import pandas as pd

# File to Load
chart_path = "charts.csv"

# Read Chart Data File and store into Pandas DataFrames
chart_data = pd.read_csv(chart_path)

#print dataframe
chart_data
```

```python
#clean data to include only necessary columns 
chart_data_reduced_columns = chart_data[["title", "rank", "date",'artist', "region", "chart", "streams"]]

#print dataframe
chart_data_reduced_columns
```

```python
# look for missing values
chart_data_reduced_columns.count()
```

```python
# drop null rows
chart_data_reduced_columns = chart_data_reduced_columns.dropna(how='any')
```

```python
# verify counts
chart_data_reduced_columns.count()
```

```python
#split date column to include only the year 
chart_data_reduced_columns['date'] = chart_data_reduced_columns['date'].apply(lambda x: x.split("-")[0])
```

```python
#print dataframe 
chart_data_reduced_columns
```

```python
#group chart_data_reduced_columns by date, title, artist and region to sort data 
new_chart = chart_data_reduced_columns.groupby(['date',"title","artist",'region'])

#sum the total number of streams for song and artists for each region 
streams = new_chart['streams'].sum()

#create new dataframe 
charts_df = pd.DataFrame({"total streams": streams,})

#print dataframe 
charts_df
```

```python
#create new csv with cleaned data 
charts_df.to_csv('spotify_charts.csv')
```
