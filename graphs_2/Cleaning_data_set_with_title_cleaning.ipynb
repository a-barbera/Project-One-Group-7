```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import demoji
import os

# File to Load
chart_path = "spotify_charts1.csv"

# Read Chart Data File and store into Pandas DataFrames
chart_data = pd.read_csv(chart_path)

#print dataframe
chart_data
```

```python
#cleaning the titles, removing characters and emojis:

#find emoji
symbol = '''ü™ê'''
demoji.findall(symbol)
```

```python
#replace emoji
cleaned_df = chart_data.replace({'\U0001fa90': 'ringed planet'})
cleaned_df['title']
```

```python
cleaned_df['title'] = cleaned_df['title'].str.strip()
cleaned_df['title'] = cleaned_df['title'].str.replace('#','')
cleaned_df['title'] = cleaned_df['title'].str.replace('"','')
cleaned_df['title'] = cleaned_df['title'].str.replace('.','')
cleaned_df['title'] = cleaned_df['title'].str.replace('$','s')
cleaned_df.replace(r'\s*(.*?)\s*', r'\1', regex=True)
cleaned_df
```

```python
#create new csv with cleaned data 
cleaned_df.to_csv('spotify_charts_clean_titles.csv', index=False)
```
